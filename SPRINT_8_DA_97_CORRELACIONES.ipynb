### __Correlación entre dos variables categóricas__. V de Cramér
---

```python
column_list_1 = ['product_category', 'payment_method']
column_list_2 = ['device_type','shipping_region']
```

```python
df[column_list_1].head()
```

```python
df[column_list_2].head()
```

```python
tabla_1 = pd.crosstab(df['product_category'], df['payment_method'])
tabla_1
```

```python
from scipy.stats import chi2_contingency
```

```python
chi2,_,_,_ = chi2_contingency(tabla_1)   #ESTADÍSTICO DE X^2
n = tabla_1.sum().sum() #FORMA FANCY CALCULAR LA CANTIDAD DE REGISTROS.
chi2
```

```python
cramers_v = np.sqrt(chi2/(n*(min(tabla_1.shape)-1)))
cramers_v
```
